# Build & Evaluate Model

---
#### **1. Train Using 4 Differents Model**
---

```python
bioresponceReady_df = bioresponce_df.copy()
```

```python
bioresponce_features = bioresponceReady_df.drop('Activity', axis=1)
bioresponce_target = bioresponceReady_df['Activity']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bioresponce_features, bioresponce_target, test_size=0.25, random_state=42)
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

common_params = {'random_state': 42, 'class_weight': 'balanced'}

# RANDOM FOREST ON SMALL TREES
small_tree_rf = RandomForestClassifier(n_estimators=250, max_depth=5, **common_params)
small_tree_rf.fit(X_train, y_train)
small_tree_rf_predictions = small_tree_rf.predict(X_test)

# RANDOM FOREST ON DEEP TREES
deep_tree_rf = RandomForestClassifier(n_estimators=250, max_depth=50, **common_params)
deep_tree_rf.fit(X_train, y_train)
deep_tree_rf_predictions = deep_tree_rf.predict(X_test)

# DEEP DECISION TREE
deep_tree = DecisionTreeClassifier(max_depth=5, **common_params)
deep_tree.fit(X_train, y_train)
deep_tree_predictions = deep_tree.predict(X_test)

# SMALL DECISION TREE
small_tree = DecisionTreeClassifier(max_depth=2, **common_params)
small_tree.fit(X_train, y_train)
small_tree_predictions = small_tree.predict(X_test)
```

---
#### **2. Calculate Using Following Metrixs**
---

```python
from sklearn.metrics import classification_report, log_loss

# DEFINE THE MODELS
models = [
    ("Random Forest (Small Trees)", small_tree_rf, small_tree_rf_predictions),
    ("Random Forest (Deep Trees)", deep_tree_rf, deep_tree_rf_predictions),
    ("Deep Decision Tree", deep_tree, deep_tree_predictions),
    ("Small Decision Tree", small_tree, small_tree_predictions)
]

for model_name, model, predictions in models:
    report = classification_report(y_test, predictions, output_dict=True)
    logloss = log_loss(y_test, model.predict_proba(X_test))

    print(f"{model_name} Classification Report:")
    print(f"Accuracy: \t{report['accuracy']:.5f}")
    print(f"Precision: \t{report['weighted avg']['precision']:.5f}")
    print(f"Recall: \t{report['weighted avg']['recall']:.5f}")
    print(f"F1-Score: \t{report['weighted avg']['f1-score']:.5f}")
    print(f"Log-Loss: \t{logloss:.5f}")
    print()
```

---
#### **3. Precision-Recall and ROC Curve**
---

```python
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc

# DEFINE THE MODELS
models = {
    "Random Forest on Small Trees": small_tree_rf,
    "Random Forest on Deep Trees": deep_tree_rf,
    "Deep Decision Tree": deep_tree,
    "Small Decision Tree": small_tree
}

plt.figure(figsize=(12, 6))

# PRECISSION-RECALL CURVE
plt.subplot(1, 2, 1)
for name, model in models.items():
    prob = model.predict_proba(X_test)[:, 1]
    precision, recall, _ = precision_recall_curve(y_test, prob)
    plt.plot(recall, precision, label=name)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.grid()

# ROC CURVE
plt.subplot(1, 2, 2)
plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Random Guess (AUC = 0.5)")
for name, model in models.items():
    prob = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, prob)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc(fpr, tpr):.2f})')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

#### **4. Train a Classifier who Avoids Type II (False Negative) Errors and Calculate the Metrics**
--- 

Type 2 Error, also known as a false negative, occurs when a test or model fails to detect something that is actually true.
In this case, I used a Random Forest Classifier and attempted to balance the data and adjust the decision threshold.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score, precision_score, accuracy_score, f1_score, log_loss

RandomForest_AvoidT2 = RandomForestClassifier(n_estimators=300, 
                                              max_depth=30, 
                                              random_state=1,
                                              class_weight='balanced')

RandomForest_AvoidT2.fit(X_train, y_train)
probabilities = RandomForest_AvoidT2.predict_proba(X_test)[:, 1]

threshold = 0.3
predictions = (probabilities >= threshold).astype(int)

recall = recall_score(y_test, predictions)
precision = precision_score(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)
f1 = f1_score(y_test, predictions)
log_loss_value = log_loss(y_test, probabilities)

print(f"Recall: \t{recall:.4f}")
print(f"Precision: \t{precision:.4f}")
print(f"Accuracy: \t{accuracy:.4f}")
print(f"F1-Score: \t{f1:.4f}")
print(f"Log-Loss: \t{log_loss_value:.4f}")
```

Here, I achieved a Recall greater than 95%.
I am unsure whether this is the right approach, since the focus is on increasing the Recall accuracy only. To achieve this, I set the threshold to 0.3. However, this adjustment caused the precision to drop.
